```python
def metropolissamp(pdfunc, get_para, Nsamp, para0, debug=False):
    """Generate Nsamp samples from pdf via Metropolis-Hastings
    
    """
    nkeep = 0
    p0 = pdfunc(para0)
    keeps = []
    for _ in itertools.repeat(None, Nsamp):
        para = get_para(para0)    # take one spect in parameter space
        p = pdfunc(para)
        chance = np.random.random()
        if debug:
            print(para, nkeep, p0, p, chance)
        if p/p0 > chance:
            keeps.append(para)
            nkeep += 1
            p0 = p
            para0 = para
        else:
            keeps.append(para0)
    return nkeep, keeps 
```

```python
v_sound = 340.29      # m/s
g = 9.81              # m/s^2

def signaltime(depth):
    return falltime(depth) + soundtime(depth)

def falltime(depth):
    return math.sqrt(2 * depth / g)

def soundtime(depth):
    return depth / v_sound
```

```python
t_sample = 0.2 * np.random.randn(10) + 3.2
```

```python
sigsq_d = np.var(t_sample)
```

```python
def pdf_stone(depth):
    misfits = [(signaltime(depth) - time)**2/sigsq_d for time in t_sample]
    return math.exp((-1)*sum(misfits)/len(t_sample))
```

```python
def getnew(depth):
    return depth + np.random.uniform(-2.0, 2.0)
```

```python
ntry, keeps = metropolissamp(pdf_stone, getnew, 1000, 20)
```
